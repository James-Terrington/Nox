```python
import os
import cv2

mask_dir = os.getcwd() + os.sep + "nox data" + os.sep
deleted = 0
for fname in os.listdir(mask_dir):
    if fname.startswith('x') and fname.endswith('.png'):
        path = os.path.join(mask_dir, fname)
        img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            os.remove(path)
            deleted += 1
            print(f"Deleted corrupt: {fname}")
print(f"Total deleted: {deleted}")
```
